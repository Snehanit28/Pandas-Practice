# Pandas Series # 
A series is a one-dimensional labeled array capable of holding any data type (integers, strings, floating point numbers, Python objects, etc.). The axis labels are collectively referred to as the index.

### 1. Creating Series ###
There are multiple ways to create a pandas series :

```python
import numpy as np
import pandas as pd
```

```python
lables = ['a', 'b', 'c']
my_list = [10,20,30]
arr = np.array([10,20,30])
d = {1:10,2:20,3:30}
```

```python
pd.Series(my_list)
```

```python
pd.Series(my_list,index=lables)
```

```python
pd.Series(arr)
```

```python
pd.Series(d)
```
