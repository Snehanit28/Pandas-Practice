# GroupBy , Aggregation #

```python
import numpy as np 
import pandas as pd
```

##### *1. GroupBy* #####

- sum(),mean(),max(),min()

```python
data = {
'Category': ['A', 'B', 'A', 'B', 'A', 'B', 'A', 'B'],
'Store': ['S1', 'S1', 'S2', 'S2', 'S1', 'S2', 'S2', 'S1'],
'Sales': [100, 200, 150, 250, 120, 180, 200, 300],
'Quantity': [10, 15, 12, 18, 8, 20, 15, 25],
'Date': pd.date_range('2023-01-01', periods=8)
}
df = pd.DataFrame(data)
```

```python
df
```

```python
cat = df.groupby('Category')
for i ,v in cat :
    print(i)
    print(v)
```

```python
# Group by 'Category' and calculate the sum of 'Sales'
df.groupby('Category')['Sales'].sum()
```

```python
# Group by 'Store' and calculate the mean of 'Quantity'
df.groupby('Store')['Quantity'].mean()
```

```python
# Group by multiple columns ('Category' and 'Store') and calculate the sum of 'Sales'
df.groupby(['Category', 'Store'])['Sales'].sum()
```

```python
# Group by multiple columns ('Category' and 'Store') and calculate the min of 'Sales'
df.groupby(['Category', 'Store'])['Sales'].min()
```

```python
# Group by 'Category' and calculate the max of 'Sales' and 'Quantity'
df.groupby('Category')[['Sales', 'Quantity']].max()
```

```python
# Group by 'Category' and calculate the max of 'Sales' and 'Quantity'
df.groupby('Category')[['Sales', 'Quantity']].sum()
```

##### *2. Aggreration* #####

```python
df
```

```python
df['Sales'].sum()    # sum gives the total of the 'Sales' column
```

```python
df['Quantity'].mean()    # mean gives the average value of the 'Quantity' column
```

```python
df['Sales'].max()    # max gives the maximum value in the 'Sales' column
```

```python
df['Quantity'].min()   # min gives the minimum value in the 'Quantity' column
```

```python
df['Sales'].median()    # median gives the middle value of the 'Sales' column when sorted
```

```python
df['Sales'].mode()    # mode gives the most frequently occurring value in the 'Sales' column
```

```python
df['Sales'].count()   # count gives the number of non-null values in the 'Sales' column
```

```python
df['Sales'].std()     # std gives the number  
```

```python
df['Sales'].agg(['mean','median','max','min'])  # here we cannot use mode 
```
