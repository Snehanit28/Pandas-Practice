# Missing Data #

```python
import numpy as np
import pandas as pd
```

##### *1. Finding Missing Data* #####

```python
data = {
    'A': [1,2,3,4,5],
    'B': [np.nan,2,3,4,5],
    'C': [1,2,3,np.nan,np.nan],
    'D': [1,2,np.nan,np.nan,5]
}
df = pd.DataFrame(data)
df
```

```python
df.isna()
```

```python
df.isna().sum()
```

```python
df.isna().any()
```

##### *2. Removing Missing Data* #####

```python
df
```

```python
df.dropna()
```

```python
df.dropna(thresh=3)
```

##### *3. Filling the missing data* #####

```python
df
```

```python
df.fillna(0)
```

```python
values = {'A':100,'B':0,'C':2,'D':9}
df.fillna(value=values)
```

```python
df.fillna(df.mean())
```
