# Data frames #

```python
import numpy as np 
import pandas as pd
```

##### *1. Creating a DataFrame* #####

```python
# from dictionary
data ={
    "Name" : ['Ram','Shyam','Amit','Aditi','Karan','Aditya','Sonia'],
    "Age" : [25,21,36,28,48,19,20],
    "City" : ['Nagpur','Mumbai','Delhi','Panjab','UP','Bihar','Kolkata'],
    "Salary":[45000,52000,60000,49000,75000,30000,35000],
    "Performance Score":[85,79,82,87,98,92,80]
}
df = pd.DataFrame(data)
df
```

```python
# From datalist
data_list = [
    ['Ram',25,'Nagpur',35000],
    ['Shyam',21,'Mumbai',52000],
    ['Amit',36,'Delhi',60000],
    ['Aditya',19,'Bihar',40000],
    ['Sonia',20,'Kolkata',45000]
]
df1 = pd.DataFrame(data_list)
df1
```

```python
heading = ['Name','Age','City','Salary']
df2 = pd.DataFrame(data_list,columns=heading)
df2
```
